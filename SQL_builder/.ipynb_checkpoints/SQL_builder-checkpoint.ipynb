```python
zeiss_visit_date = recent[recent['modality']=='structural_oct_6mmx6mm']['visit_date'].to_list()
list_date = recent['visit_date'].to_list()
actual_exams = [_ in zeiss_visit_date for _ in list_date]
real = recent.loc[actual_exams]
```

```python
oct = real[real['modality']=='oct']
wrong_nb_imgs = oct[oct['nb_imgs']!=96]
wrong_nb_imgs.sort_values('id', inplace=True)
for index, value  in wrong_nb_imgs.iterrows():
    print(value['id'], value['surname'], value['last_name'], value['visit_date'], value['eye'])
```

```python
oct = real[real['modality']=='oct']
good_nb_imgs = oct[oct['nb_imgs']==96]
good_nb_imgs.sort_values('id', inplace=True)

for index, value  in good_nb_imgs.iterrows():
    print(value['id'], value['surname'], value['last_name'], value['visit_date'], value['eye'])
```
